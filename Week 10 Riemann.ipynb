```python
#Left Hand Riemann
import numpy as np
import matplotlib.pyplot as plt

def default_f(x):
    return np.sqrt(1 + x**3)

def Riemann(a, b, n, fx = default_f , method = 'c', graphics = True):
    deltax = (b-a)/n
    
    #for bar plotting arguments, must be negative for right align
    w_mult = 1
    
    #method to define placement of sums.
    if method == 'l':
        x = a
        align1 = 'edge'
    elif method == 'c':
        x = a + deltax/2
        align1 = 'center'
    elif method == 'r':
        x = a + deltax
        align1 = 'edge'
        w_mult = -1
        
    #initialize accumulator- sum of all boxes.
    acc = 0
    if graphics == True:
        fig, ax = plt.subplots()
        X = np.linspace(a, b, n)
        ax.plot(X, fx(X), color = 'limegreen')
        
    for i in range(n):
        #optional plotting
        if graphics == True:
            ax.bar(x, fx(x), width = (w_mult * deltax), align = align1, color = 'hotpink', ec = "hotpink", linewidth = 2, alpha = 0.5)
        deltaS = fx(x) * deltax
        acc = acc + deltaS
        x = x + deltax
    if graphics == True:
        plt.annotate(f'Total ={acc:.3f}', (a, fx(x)/2))
        plt.show()
        
    return acc
        
        
    
    
```

```python
#with graphics
Riemann(1, 6, 8, method = 'l')

#without graphics 
sum_1 = Riemann(1, 6, 8, method = 'l', graphics = False)
print("The sum is ", sum_1)
```

```python
#center aligned
Riemann(1, 6, 8)
```

```python
#right aligned
Riemann(1, 6, 8, method = 'r')

#right aligned more bins
Riemann(1, 6, 16, method = 'r')
```

## Activity 4:
Activity #4 Modify RIEMANN again with either the Trapezoid rule or the Simpsons Rule, section 11.6  What are the advantages of these methods?

- _The advantage of both the Trapezoid and Simpson's Rule is that they reduce error in the calculation of area for each interval under the curve. The trapezoid rule changes the usual rectangle approximation to a trapezoid, which computes the area as the average of the areas of both rectangles. However, the midpoint approximation is still better, so the Simpsons rule improves upon the trapezoid rule by taking the averae of 2 midpoint volumes and one trapezoidal volume calculated over the same interval, which further reduces the error. _

```python
def TzRiemann(a, b, n, fx = default_f, graphics = True):
    deltax = (b-a)/n
    
    #for bar plotting arguments, must be negative for right align
    w_mult = 1
    
    #calculate both left and right.

    x1 = a
    align1 = 'edge'
    
    x2 = a + deltax
    w_mult = -1
        
    #initialize accumulator- sum of all boxes.
    acc = 0
    if graphics == True:
        fig, ax = plt.subplots()
        X = np.linspace(a, b, n)
        ax.plot(X, fx(X), color = 'limegreen')
        
    for i in range(n):
        #optional plotting
        if graphics == True:
            ax.bar(x1, fx(x1), width = (deltax), align = 'edge', color = 'hotpink', ec = "hotpink", linewidth = 2, alpha = 0.5)
            ax.bar(x2, fx(x2), width = (-deltax), align = 'edge', color = 'turquoise', ec = "turquoise", linewidth = 2, alpha = 0.5)
            
        deltaS1 = fx(x1) * deltax
        deltaS2 = fx(x2) * deltax
        
        #trapezoidal approximation of each rectangle from left and right sums
        tv = (deltaS1 + deltaS2)/2
        
        acc = acc + tv
        x1 = x1 + deltax
        x2 = x2 + deltax
    if graphics == True:
        plt.annotate(f'Total ={acc:.3f}', (a, fx(x1)/2))
        plt.show()
        
    return acc
```
